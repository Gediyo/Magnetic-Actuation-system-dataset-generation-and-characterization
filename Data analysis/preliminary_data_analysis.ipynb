```python
import pandas as pd
import datetime
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import numpy as np
```

```python
# Load the CSV file into a DataFrame
df = pd.read_csv('24-07-2024_20-02-49_Coil_table_data.csv', index_col=0)

df.index = pd.to_datetime(df.index, format="%d-%m-%Y_%H-%M-%S")

df = df.resample('s').mean()

# Calculate the 3-period running average
df['Running_Avg_coilTemp_5'] = df['msdCoilTemp_5(degC)'].rolling(window=10).mean()

df['Running_Avg_coilTemp_1'] = df['msdCoilTemp_1(degC)'].rolling(window=10).mean()

df['Running_Avg_coilTemp_3'] = df['msdCoilTemp_3(degC)'].rolling(window=10).mean()
df['Running_Avg_coilTemp_4'] = df['msdCoilTemp_4(degC)'].rolling(window=10).mean()
df['Running_Avg_coilTemp_7'] = df['msdCoilTemp_7(degC)'].rolling(window=10).mean()
df['Running_Avg_coilTemp_8'] = df['msdCoilTemp_8(degC)'].rolling(window=10).mean()
```

```python
df13 = df[df["cmdCoilCurrent_1(A)"]>=13]
```

```python
print(df13['msdCoilCurrent_1(A)'].max())
print(df13['msdCoilCurrent_1(A)'].min())
print(df13['msdCoilCurrent_1(A)'].mean())
print((df13['msdCoilCurrent_1(A)'].max()- df13['msdCoilCurrent_1(A)'].min())/2)
```

```python
fig, ax1 = plt.subplots(figsize=(10, 6))

ax1.plot(df.index, df['cmdCoilCurrent_5(A)'],  linestyle='-', color='b')
ax1.set_xlabel('Time (Hr:min)', fontsize=16)
ax1.set_ylabel('Command current [A]', color='b', fontsize=16)
ax1.tick_params(axis='y', labelcolor='b', labelsize=12)
ax1.tick_params(axis='x', rotation=45, labelsize=12)

# Format the x-axis to show only hour and minute
ax1.xaxis.set_major_formatter(mdates.DateFormatter('%H:%M'))
#ax1.xaxis.set_major_locator(mdates.HourLocator(interval=1))  # Adjust the interval as needed

# Create a secondary y-axis
ax2 = ax1.twinx()  # This creates a new Axes instance sharing the same x-axis
ax2.plot(df.index, df['Running_Avg_coilTemp_4'], linestyle='-', color='r')
ax2.set_ylabel('Running avg. coil temperature [degC]', color='r', fontsize=16)
ax2.tick_params(axis='y', labelcolor='r', labelsize=12) 

plt.title('Running Average coil temperature relative to command current  (Coil #5)', fontsize=18)
plt.grid(True)
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability

# Optional: add legends
ax1.legend(loc='upper left')
ax2.legend(loc='upper right')

plt.tight_layout()  # Adjust layout to make sure everything fits
plt.show()
```
